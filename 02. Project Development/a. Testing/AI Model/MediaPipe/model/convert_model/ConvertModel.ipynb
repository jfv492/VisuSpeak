```python
import tensorflow as tf
from tensorflow import keras
from tensorflowjs import converters

# Load your Keras model
model = keras.models.load_model('/Users/jasmeetsingh/Desktop/Capstone/AI Model Testing/hand-gesture-recognition-mediapipe/model/keypoint_classifier/keypoint_classifier.hdf5')

# Save the model to the TensorFlow SavedModel format
model_save_path = '/Users/jasmeetsingh/Desktop/Capstone/AI Model Testing/hand-gesture-recognition-mediapipe/model/convert_model'
model.save(model_save_path, save_format='tf')

# Convert the SavedModel to TensorFlow.js format
tfjs_target_dir = '/Users/jasmeetsingh/Desktop/Capstone/AI Model Testing/hand-gesture-recognition-mediapipe/model/convert_model'
converters.convert_tf_saved_model(model_save_path, tfjs_target_dir)
```
