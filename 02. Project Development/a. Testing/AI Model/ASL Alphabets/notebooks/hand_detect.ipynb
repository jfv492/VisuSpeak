```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
face_cascade = cv2.CascadeClassifier('/Users/Belal/opencv/data/haarcascades/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('/Users/Belal/opencv/data/haarcascades/haarcascade_eye.xml')

smile_cascade = cv2.CascadeClassifier('/Users/Belal/opencv/data/haarcascades/haarcascade_smile.xml')
```

```python
#img = cv2.imread("/Users/Belal/Desktop/dat_face.jpg")
img = cv2.imread('../sample_data/')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
type(img)
```

```python
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
```
